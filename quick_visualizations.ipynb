```python
import pandas as pd
df = pd.read_csv("/content/ds2002project/college_data_clean.csv")
```

```python
df.head()
```

```python
# Seaborn plot
# Import Seaborn
# Import Seaborn and Matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix using only numeric columns
numeric_tips = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_tips.corr()

# Create a heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Seaborn Heatmap of Numeric Features in College Dataset')
plt.show()
```

```python
# Import Matplotlib
import matplotlib.pyplot as plt
va_colleges = df.loc[df['state'] == 'VA']
va_colleges.head()

fig, ax = plt.subplots()


ax.bar(va_colleges['name'], va_colleges['earnings'])


ax.set_ylabel('Earnings')
ax.set_xlabel('College')
ax.set_title('Average Earnings After Graduation of VA Colleges')
plt.xticks(rotation=90, fontsize=5.5)

plt.show()
```
